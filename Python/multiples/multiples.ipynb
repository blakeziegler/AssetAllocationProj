```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load the data
pe_quarterly = pd.read_csv("pe_ratio_clean_total.csv")

# Convert 'Date' column to datetime if not already done
pe_quarterly['Date'] = pd.to_datetime(pe_quarterly['Date'], format='%Y-%m-%d')

# Filter the data for the specified range (1948-2024)
start_year = 1948
end_year = 2024
filtered_data = pe_quarterly[(pe_quarterly['Date'].dt.year >= start_year) & 
                             (pe_quarterly['Date'].dt.year <= end_year)].copy()

# Define a function to categorize into 3-year intervals
def categorize_year(date):
    year = date.year
    return f"{year // 3 * 3}-{year // 3 * 3 + 2}"

# Apply the function to create a new column for 3-year intervals
filtered_data['3_year_interval'] = filtered_data['Date'].apply(categorize_year)

# Calculate the average for each 3-year interval
averages_3_years = filtered_data.groupby('3_year_interval')['value'].mean().reset_index()

# Convert the '3_year_interval' to a suitable format for plotting
averages_3_years['start_year'] = averages_3_years['3_year_interval'].apply(lambda x: int(x.split('-')[0]))

# Linear Regression
X = averages_3_years['start_year'].values.reshape(-1, 1)
y = averages_3_years['value'].values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Calculate R-squared
r_squared = r2_score(y, y_pred)

# Extend the dataset with future years (10 years into the future)
future_years = list(range(2025, 2035))
future_X = np.array(future_years).reshape(-1, 1)
future_y_pred = model.predict(future_X)

# Combine current and future data for plotting
all_years = np.concatenate([X.flatten(), future_X.flatten()])
all_predictions = np.concatenate([y_pred, future_y_pred])

# Plot the 3-year interval averages
plt.figure(figsize=(14, 7))
plt.plot(averages_3_years['start_year'], averages_3_years['value'], marker='o', linestyle='-', color='teal', label='P/E Ratio')

# Add labels and title
plt.title('Average P/E Ratios Over 3-Year Intervals (1948-2034)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average P/E Ratio', fontsize=14)

# Enhance gridlines
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Plot the combined regression line
plt.plot(all_years, all_predictions, color='red', linestyle='--', label=f'Linear Regression (R = {r_squared:.2f})')

# Set x-axis ticks every 6 years
plt.xticks(np.arange(start_year, 2035, 6))

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

# Print the regression coefficients
print("Linear Regression Coefficients:")
print(f"Slope: {model.coef_[0]:.2f}")
print(f"R: {r_squared:.2f}")

# Print the forecasted P/E ratios
forecasted_pe_ratios = pd.DataFrame({'Year': future_years, 'Forecasted P/E Ratio': future_y_pred})
print("\nForecasted P/E Ratios (2025-2034):")
print(forecasted_pe_ratios)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load the data
pe_quarterly = pd.read_csv("pe_ratio_clean_total.csv")

# Convert 'Date' column to datetime if not already done
pe_quarterly['Date'] = pd.to_datetime(pe_quarterly['Date'], format='%Y-%m-%d')

# Filter the data for the specified range (2010-2024)
start_year = 2004
end_year = 2024
filtered_data = pe_quarterly[(pe_quarterly['Date'].dt.year >= start_year) & 
                             (pe_quarterly['Date'].dt.year <= end_year)].copy()

# Define a function to categorize into quarters
def categorize_quarter(date):
    year = date.year
    quarter = (date.month - 1) // 3 + 1
    return f"{year}-Q{quarter}"

# Apply the function to create a new column for quarters
filtered_data['quarter'] = filtered_data['Date'].apply(categorize_quarter)

# Calculate the average for each quarter
averages_quarter = filtered_data.groupby('quarter')['value'].mean().reset_index()

# Convert the 'quarter' to a suitable format for plotting
averages_quarter['year_quarter'] = averages_quarter['quarter'].apply(lambda x: int(x.split('-')[0]) + (int(x.split('-')[1][1]) - 1) / 4)

# Linear Regression
X = averages_quarter['year_quarter'].values.reshape(-1, 1)
y = averages_quarter['value'].values
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Calculate R-squared
r_squared = r2_score(y, y_pred)

# Extend the dataset with future quarters (10 years into the future, 40 quarters)
future_quarters = list(range(2025, 2028))
future_quarters = [f"{year}-Q{quarter}" for year in future_quarters for quarter in range(1, 5)]
future_year_quarters = np.array([int(fq.split('-')[0]) + (int(fq.split('-')[1][1]) - 1) / 4 for fq in future_quarters]).reshape(-1, 1)
future_y_pred = model.predict(future_year_quarters)

# Combine current and future data for plotting
all_year_quarters = np.concatenate([X.flatten(), future_year_quarters.flatten()])
all_predictions = np.concatenate([y_pred, future_y_pred])

# Plot the quarterly averages
plt.figure(figsize=(14, 7))
plt.plot(averages_quarter['year_quarter'], averages_quarter['value'], marker='o', linestyle='-', color='teal', label='P/E Ratio')

# Add labels and title
plt.title('Average P/E Ratios by Quarter (2010-2024)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average P/E Ratio', fontsize=14)

# Enhance gridlines
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Plot the combined regression line
plt.plot(all_year_quarters, all_predictions, color='red', linestyle='--', label=f'Linear Regression (R = {r_squared:.2f})')

# Set x-axis ticks every 2 years
plt.xticks(np.arange(start_year, 2028, 2))

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

# Print the regression coefficients
print("Linear Regression Coefficients:")
print(f"Slope: {model.coef_[0]:.2f}")
print(f"R: {r_squared:.2f}")

# Print the forecasted P/E ratios
forecasted_pe_ratios = pd.DataFrame({'Year_Quarter': future_quarters, 'Forecasted P/E Ratio': future_y_pred})
print("\nForecasted P/E Ratios (2025-2034):")
print(forecasted_pe_ratios)
```
