```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
pe_quarterly = pd.read_csv("pe_ratio_clean_total.csv")

# Convert 'Date' columns to datetime if not already done
pe_quarterly['Date'] = pd.to_datetime(pe_quarterly['Date'])

# Plot the quarterly P/E ratios
plt.figure(figsize=(14, 7))

# Plot with improved aesthetics
plt.plot(pe_quarterly['Date'], pe_quarterly['value'], label='Quarterly P/E Ratio', color='teal', linewidth=2)

# Set x-axis ticks to every 10 years
plt.xticks(pd.date_range(start=pe_quarterly['Date'].min(), end=pe_quarterly['Date'].max(), freq='10Y'), rotation=45)

# Remove x-axis gridlines and enhance y-axis gridlines
plt.grid(axis='x', which='both', linestyle='', linewidth=0)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Add labels and title
plt.title('Quarterly P/E Ratios Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('P/E Ratio', fontsize=14)

# Add a legend with improved styling
plt.legend(loc='upper left', fontsize=12, frameon=True, framealpha=0.9, shadow=True, borderpad=1)

# Add a background color
plt.gca().set_facecolor('#f7f7f7')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
pe_quarterly = pd.read_csv("pe_ratio_clean_total.csv")

# Convert 'Date' columns to datetime if not already done
pe_quarterly['Date'] = pd.to_datetime(pe_quarterly['Date'])

# Normalize the 'value' column using a logarithmic transformation
pe_quarterly['log_value'] = np.log(pe_quarterly['value'])

# Define the recession periods
recession_periods = [
    ('1945-10-01', '1948-10-01'),
    ('1969-12-01', '1972-12-01'),
    ('1997-02-01', '2000-02-01'),
    ('2005-11-01', '2008-11-01'),
    ('2017-01-01', '2020-01-01')
]

# Plot the normalized quarterly P/E ratios
plt.figure(figsize=(14, 7))

# Add background shading for recession periods
for start_date, end_date in recession_periods:
    plt.axvspan(pd.to_datetime(start_date), pd.to_datetime(end_date), color='red', alpha=0.3)

# Plot the data
plt.plot(pe_quarterly['Date'], pe_quarterly['log_value'], label='Quarterly P/E Ratio (Log Normalized)', color='teal', linewidth=2)

# Set x-axis ticks to every 10 years on January 1st
start_year = pe_quarterly['Date'].dt.year.min()
end_year = pe_quarterly['Date'].dt.year.max()
x_ticks = pd.date_range(start=f'{start_year}-01-01', end=f'{end_year}-01-01', freq='5Y')

plt.xticks(x_ticks, rotation=45)

# Remove x-axis gridlines and enhance y-axis gridlines
plt.grid(axis='x', which='both', linestyle='', linewidth=0)
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Add labels and title
plt.title('Pre-Recession P/E Ratios', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Log Normalized P/E Ratio', fontsize=14)

# Add a legend with improved styling
plt.legend(loc='upper left', fontsize=12, frameon=True, framealpha=0.9, shadow=True, borderpad=1)

# Add a background color
plt.gca().set_facecolor('#f7f7f7')

# Show the plot
plt.tight_layout()
plt.show()

# Analyze data over these time periods
# Analyze current and historic large cap PE to find a similar 3.0 and 3.5 celling
```
